```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/melb_data.csv")
```

```python
#The first column contains string ids from which you need to extract only the numerical part and convert them to integers.
import re
df['Unique ID']
def find_number(text):
  numbers=re.findall(r'[0-9]+',text)
  return " ".join(numbers)
df['numbers_from_string']=df['Unique ID'].apply(lambda x: find_number(x))
df['numbers_from_string']
```

```python
#Next we will use those extracted numbers and set them as the primary index. We can achieve this using the set_index function.
df.set_index('numbers_from_string', inplace = True)
```

```python
df
```

```python
#Deal with NaN Values if there are any. There can be two ways, you can either delete those entries having missing values or you can put in values yourself with some methods e.g. mean.

column_means=df.median()
df.fillna(column_means,inplace=True)
df
```

```python
#Try memory optimization. You can use methods like astype() to cast data types like float to int.

df['Distance']=df['Distance'].astype(int)
df
```

```python
#Use panda functions to calculate mean, median, mode, variance of columns like ‘rooms’ and ‘prices’ to understand the dataset better.

x=df['Price'].mean()
y=df['Rooms'].median()
z=df['Propertycount'].mode()
p=df['BuildingArea'].max()
print('Mean of prices :',x)
print('Median of rooms: ',y)
print('Mode of propertycount :',z)
print('Maximum building area : ',p)
```
